```python
# Load the dataset
from google.colab import files
uploaded = files.upload()
# Function to load CSV file using Pandas
import pandas as pd
dataset = pd.read_csv('sales_data.csv')
dataset.head()
```

```python
# Check the column names in the DataFrame
print(dataset.columns)

# After identifying the correct column name, which is 'Sales_Amount' based on the output
# Use the correct variable name 'dataset' instead of 'df'
# Replace 'Total_Sales' with the actual column name from your dataset
correct_sales_column_name = 'Sales_Amount' # Changed to the correct column name

if correct_sales_column_name in dataset.columns:
    dataset['Sales_Tax'] = dataset[correct_sales_column_name] * 0.10  # 10% tax
    dataset['Total_With_Tax'] = dataset[correct_sales_column_name] + dataset['Sales_Tax']

    # Preview updated DataFrame
    display(dataset.head())
else:
    # This else block should not be reached now that the column name is correct
    print(f"Error: Column '{correct_sales_column_name}' not found in the dataset.")
    print("Available columns:", dataset.columns.tolist())
```

```python
# Export to new CSV
dataset.to_csv('updated_sales_data.csv', index=False)

# Download the file in Colab
files.download('updated_sales_data.csv')
```
