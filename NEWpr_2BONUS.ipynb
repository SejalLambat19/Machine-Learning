```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd

# Ensure that 'titanic.csv' is present in the '/content/' directory,
# or update the path below to the correct location of the file.
Titanic = "/content/titanic.csv"
titanic_df = pd.read_csv(Titanic)

# Survival rate by class and gender
sns.barplot(x="Pclass", y="Survived", hue="Sex", data=titanic_df)
plt.title("Survival Rate by Class and Gender")
plt.show()

# Age distribution
sns.histplot(titanic_df['Age'], kde=True, bins=20)
plt.title("Age Distribution of Passengers")
plt.xlabel("Age")
plt.ylabel("Frequency")
plt.show()
```
